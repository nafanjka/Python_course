```python
# to add a new rows we need to trnspose able first
df1_t=df1.T # T - transpose
df1_t
```

```python
# now we can add new record to transposed table
df1_t["My Address"]=[7,"My City","My State","My Country","My Shop",8,"My Continent"]
df1_t
```

```python
# Now we can transpose it back and assign to our main variabe
df1=df1_t.T
df1
```
