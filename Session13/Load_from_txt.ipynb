```python
# for comma-separated file we can use read_csv
import pandas
df4=pandas.read_csv("supermarkets-commas.txt")
df4
```

```python
# if you have data_sheet without headers we can specify that header is missing then it will add default header
df6=pandas.read_csv("supermarkets-semi-colons.txt", sep=";", header=None)
df6
```

```python
# also we can assign the own column names to this non-header data
df6.columns=("ID", "Address", "City", "State", "Country", "Name", "Employees")
df6
```
